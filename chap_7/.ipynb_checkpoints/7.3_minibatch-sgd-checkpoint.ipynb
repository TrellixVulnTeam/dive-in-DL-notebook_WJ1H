```python
###　梯度下降（批量梯度下降）：（在每一次迭代中）使用整个训练集求梯度;

### 随机梯度下降：　随机采样一个样本计算梯度;

### 小批量随机梯度下降：先随机均匀采样多个样本来组成一个小批量，后使用这个小批量来计算梯度
```

小批量随机梯度下降算法中的梯度求值

$$ \boldsymbol{g}_t \leftarrow \nabla f_{\mathcal{B}_t}(\boldsymbol{x}_{t-1}) = \frac{1}{|\mathcal{B}|} \sum_{i \in \mathcal{B}_t}\nabla f_i(\boldsymbol{x}_{t-1}) $$

小批量随机梯度下降算法中的梯度更新

$$\boldsymbol{x}_t \leftarrow \boldsymbol{x}_{t-1} - \eta_t \boldsymbol{g}_t.$$

关于学习率需要进行自我衰减．

* 为什么？因为基于随机采样得到的梯度的方差在迭代过程中无法减小．

* 实施方法，令$\eta_t = \eta t^\alpha \left( 通常\alpha=-1 \ or \ \alpha=-0.5 \right), \eta_t = \eta \alpha^t \left( 通常\alpha=0.95\right)$, 这样，学习率和随机梯度乘积的方差会减小．

小批量随机梯度下降中每次迭代的计算开销为$\mathcal{O}(|\mathcal{B}|)$。当批量大小为1时，该算法即为随机梯度下降；当批量大小等于训练数据样本数时，该算法即为梯度下降。

```python
import numpy as np
import time
import torch
from torch import nn, optim
import sys
sys.path.append('../d2lzh/')
import d2lzh_pytorch as d2l

def get_data_ch7():
    data = np.genfromtxt('./airfoil_self_noise.dat', delimiter='\t')
    data = (data - data.mean(axis=0)) / data.std(axis=0)
    
    return torch.tensor(data[:1500, :-1], dtype=torch.float32),\
        torch.tensor(data[:1500, -1], dtype=torch.float32)
                        # 共1500个样本，每个样本5个特征
features, labels = get_data_ch7()
print(features.shape)
print(labels.shape)
```

```python
### 从零开始实现
```

```python
def sgd(params, states, hyperparams):
    for p in params:
        p.data -= hyperparams['lr'] * p.grad.data
```

```python
# def train_ch7(optimizer_fn, states, hyperparams, features, labels,
#              batch_size=10, num_epochs=2):
#     net, loss = d2l.linreg, d2l.squared_loss
    
#     w = torch.nn.Parameter(
#         torch.tensor(np.random.normal(0, 0.01, size=(features.shape[1],1)), 
#                      dtype=torch.float32),
#         requires_grad=True,
#     )
    
#     b = torch.nn.Parameter(
#         torch.zeros(1, dtype=torch.float32), requires_grad=True,
#     )
    
#     def eval_loss():
#         return loss(net(features, w, b), labels).mean().item()
    
#     ls = [eval_loss()]
#     data_iter = torch.utils.data.DataLoader(
#         torch.utils.data.TensorDataset(features, labels),
#         batch_size, shuffle=True,
#     )
    
#     for _ in range(num_epochs):
#         start_time = time.time()
#         for batch_i, (X, y) in enumerate(data_iter):
#             l = loss(net(X, w, b), y).mean()
            
#             if w.grad is not None:
#                 w.grad.data.zero_()
#                 b.grad.data.zero_()
#             l.bachward()
#             optimizer_fn([w,b], states, hyperparams)
#             if (batch_i + 1) * batch_size % 100 == 0:
#                 ls.append(eval_loss())
                
#     print("loss: %f, %f sec per epoch" % (ls[-1], time.time() - start_time))
#     d2l.set_figsize()
#     d2l.plt.plot(np.linspace(0, num_epochs, len(ls)), ls)
#     d2l.plt.xlabel('epoch')
#     d2l.plt.ylabel('loss')
#     d2l.plt.show()

# 本函数已保存在d2lzh_pytorch包中方便以后使用
def train_ch7(optimizer_fn, states, hyperparams, features, labels,
              batch_size=10, num_epochs=2):
    # 初始化模型
    net, loss = d2l.linreg, d2l.squared_loss
    
    w = torch.nn.Parameter(torch.tensor(np.random.normal(0, 0.01, size=(features.shape[1], 1)), dtype=torch.float32),
                           requires_grad=True)
    b = torch.nn.Parameter(torch.zeros(1, dtype=torch.float32), requires_grad=True)

    def eval_loss():
        return loss(net(features, w, b), labels).mean().item()

    ls = [eval_loss()]
    data_iter = torch.utils.data.DataLoader(
        torch.utils.data.TensorDataset(features, labels), batch_size, shuffle=True)
    
    for _ in range(num_epochs):
        start = time.time()
        for batch_i, (X, y) in enumerate(data_iter):
            l = loss(net(X, w, b), y).mean()  # 使用平均损失
            
            # 梯度清零
            if w.grad is not None:
                w.grad.data.zero_()
                b.grad.data.zero_()
                
            l.backward()
            optimizer_fn([w, b], states, hyperparams)  # 迭代模型参数
            if (batch_i + 1) * batch_size % 100 == 0:
                ls.append(eval_loss())  # 每100个样本记录下当前训练误差
    # 打印结果和作图
    print('loss: %f, %f sec per epoch' % (ls[-1], time.time() - start))
    d2l.set_figsize()
    d2l.plt.plot(np.linspace(0, num_epochs, len(ls)), ls)
    d2l.plt.xlabel('epoch')
    d2l.plt.ylabel('loss')
    d2l.plt.show()
```

```python
def train_sgd(lr, batch_size, num_epochs=2):
    train_ch7(sgd, None, {'lr':lr}, features, labels, batch_size, num_epochs)
    
train_sgd(1, 1500, 6)  # 梯度下降
```

```python
train_sgd(0.005, 1) # 随机梯度下降
```

```python
train_sgd(0.05, 10) # 小批量（１０）随机梯度下降
```

```python
### PyTorch简洁实现
```

```python
def train_pytorch_ch7(optimizer_fn, optimizer_hyperparams, 
                      features, labels, batch_size=10, num_epochs=2):
    net = nn.Sequential(
        nn.Linear(features.shape[-1], 1)
    )
    
    loss = nn.MSELoss()
    optimizer = optimizer_fn(net.parameters(), **optimizer_hyperparams)
    
    def eval_loss():
        return loss(net(features).view(-1), labels).item() / 2
    
    ls = [eval_loss()]
    data_iter = torch.utils.data.DataLoader(
        torch.utils.data.TensorDataset(features, labels), 
        batch_size, shuffle=True,
    )
    
    for _ in range(num_epochs):
        start_time = time.time()
        for batch_i, (X, y) in enumerate(data_iter):
            l = loss(net(X).view(-1), y) / 2
            
            optimizer.zero_grad()
            l.backward()
            optimizer.step()
            if (batch_i + 1) * batch_size % 100 == 0:
                ls.append(eval_loss())
            
    # 打印结果和作图
    print('loss: %f, %f sec per epoch' % (ls[-1], time.time() - start_time))
    d2l.set_figsize()
    d2l.plt.plot(np.linspace(0, num_epochs, len(ls)), ls)
    d2l.plt.xlabel('epoch')
    d2l.plt.ylabel('loss')
    d2l.plt.show()    
```

```python
train_pytorch_ch7(optim.SGD, {'lr':0.05}, features, labels, 10)
```

总结：
小批量随机梯度每次随机均匀采样一个小批量的训练样本来计算梯度。

在实际中，（小批量）随机梯度下降的学习率可以在迭代过程中进行自我衰减．

miniSGD迭代周期的耗时介于ＳＧＤ和ＧＤ之间．
