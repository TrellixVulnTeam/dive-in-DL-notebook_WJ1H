```python
# 梯度下降的问题
```

```python
import sys
sys.path.append('../d2lzh/')
import d2lzh_pytorch as d2l
import torch

eta = 0.4  # learning rate

def f_2d(x1, x2):
    return 0.1 * x1 ** 2 + 2 * x2 ** 2

def gd_2d(x1, x2, s1, s2):
    return (x1 - eta * 0.2 * x1, x2 - eta * 4 * x2, 0, 0.4)

d2l.show_trace_2d(f_2d, d2l.train_2d(gd_2d))
```

```python
# 从以上结果可以看到，在x2轴方向上的斜率比在x1轴方向上的斜率的绝对值更大；
# 也就是说，给定学习率，梯度下降法迭代自变量时会使自变量在垂直方向的移动幅度更大．
# 也就是说，我们要减小学习率来避免在垂直方向的自变量越过目标函数的最优解．
# 也就是说，这会使自变量在水平方向上朝最优解的移动速度变慢．

# 以下展示当学习率取值过大时，在垂直方向，自变量越过目标函数的最优解（发散）
```

```python
eta = 0.6  # 增大学习率b
d2l.show_trace_2d(f_2d, d2l.train_2d(gd_2d))
```

```python
### 指数加权移动平均
```

当前时间步$t$的变量$y_t$是上一时间步$t-1$的变量$y_{t-1}$和当前时间步另一变量$x_t$的线性组合：

$$ y_t = \gamma y_{t-1} + \left( 1-\gamma \right) x_t$$

展开$y_t$，有:

$$ \begin{aligned} y_t &= (1-\gamma) x_t + \gamma y_{t-1} \end{aligned} $$


$$ \begin{aligned} y_t 
&= (1-\gamma)x_t + (1-\gamma) \cdot \gamma x_{t-1} + \gamma^2y_{t-2}\  \end{aligned} $$

$$ \begin{aligned} y_t &= (1-\gamma)x_t + (1-\gamma) \cdot \gamma x_{t-1} + (1-\gamma) \cdot \gamma^2x_{t-2} + \gamma^3y_{t-3}\ \end{aligned} $$

$$ \ldots $$

在实际中，常常将$y_t$看作是对最近$1/\left( 1-\gamma \right)$个时间步的$x_t$值的加权平均．（而且距离当前时间步越近，其权值也更大）. 那么这个$1/\left( 1-\gamma \right)$是怎么推导出来的？

令$n=\frac{1}{1-\gamma}$，则有$\gamma=1-\frac{1}{n}$，$\gamma^n=\left( 1-\frac{1}{n}\right)^n = \gamma ^{\frac{1}{1-\gamma}}$，因为

$$ \lim_{n \rightarrow \infty} \left(1-\frac{1}{n}\right)^n = \exp(-1) \approx 0.3679,$$

即当$\gamma \rightarrow 1$时，有$\gamma ^n=\gamma ^{\frac{1}{1-\gamma}}=\exp{(1)}\approx 0.3679$，如果把$\exp(-1)$当作一个比较小的数，我们可以在近似中忽略所有含$\gamma^{1/(1-\gamma)}$和比$\gamma^{1/(1-\gamma)}$更高阶的系数的项．

```python
### 由指数加权移动平均　到　动量法
```

动量法：动量法是为了解决梯度下降的问题而提出的（学习率较小时，能保证解不超出最优解的范围，但学习过程相对缓慢;　学习率过大，发散）

动量法的表示：

$$  \boldsymbol{v}_t \leftarrow \gamma \boldsymbol{v}_{t-1} + \eta_t \boldsymbol{g}_t, $$

$$  \boldsymbol{x}_t \leftarrow \boldsymbol{x}_{t-1} - \boldsymbol{v}_t $$

由式子可以看出，动量法是在梯度下降法的基础上增加了一个变量$\gamma \boldsymbol{v}_{t-1}$，且$0\le \gamma \lt 1$，当$\gamma = 0$时，变回SGD.

** 如何将动量法中的$\boldsymbol{v}_t$与指数加权移动平均联系起来？**

对$\boldsymbol{v}_t$做如下变形，

$$ \boldsymbol{v}_t \leftarrow \gamma \boldsymbol{v}_{t-1} + \left( 1-\gamma \right)\left( \frac{\eta_t}{1-\gamma}\boldsymbol{g}_t \right)， $$即上面的$x_t$用$\frac{\eta_t}{1-\gamma}\boldsymbol{g}_t$来代替．直接的理解就是，$\boldsymbol{v}_t$将$1/(1-\gamma)$个$ \eta_t \boldsymbol{g}_t$（即，学习率$\times$梯度）通过指数加权移动平均的方式组合起来（然后再除以$1-\gamma$），且越靠近当前时间步的$\eta_t \boldsymbol{g}_t$所分配的权重越大．

也就是说，自变量在作出移动方向和移动幅度的选择时，不仅需要考虑当前梯度方向，还需要考虑接下来的移动方向是否与过去的移动方向保持总体一致．

```python
### 从零开始实现．
```

```python
features, labels = d2l.get_data_ch7()

def init_momentum_states():
    v_w = torch.zeros((features.shape[1], 1), dtype=torch.float32)
    v_b = torch.zeros(1, dtype=torch.float32)
    
    return (v_w, v_b)

def sgd_momentum(params, states, hyperparams):
    for p, v in zip(params, states):
        v.data = hyperparams['momentum'] * v.data + hyperparams['lr'] * p.grad.data
        p.data -= v.data
```

```python
# 将动量超参数momentum设0.5，这时可以看成是特殊的小批量随机梯度下降：
# 其小批量随机梯度为最近2个时间步的2倍小批量梯度的加权平均。
d2l.train_ch7(sgd_momentum, init_momentum_states(), 
             {'lr':0.02, 'momentum':0.5}, features, labels)
```

```python
# 将动量增大，依然收敛，但是后期变得不够稳定．
# 依然可以看成是特殊的小批量随机梯度下降：
# 其小批量随机梯度为最近10个时间步的10倍小批量梯度的加权平均。
d2l.train_ch7(sgd_momentum, init_momentum_states(),
             {'lr':0.02, 'momentum':0.9}, features, labels)
```

```python
# 为了让它保持稳定，改变学习率为原来的2/10=1/5
d2l.train_ch7(sgd_momentum, init_momentum_states(),
             {'lr':0.004, 'momentum':0.9}, features, labels)
```
