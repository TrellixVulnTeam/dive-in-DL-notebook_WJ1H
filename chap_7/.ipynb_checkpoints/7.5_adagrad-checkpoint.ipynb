```python
### AdaGrad算法（adaptive subgradients）

# 问题的引入：

# 在梯度下降算法中，各自变量（维度）梯度变化的大小可能不一致，
# 为了同时兼顾抖动最厉害的和相对平缓的几个维度，我们通常要将学习率调低，使自变量的抖动不至于过大

# 对此，动量法通过引入指数加权移动平均来考虑他们先前阶段的梯度变化方向，
# 从而约束了抖动过度的维度，同时也保证（加快）了其他维度的自变量往正确的方向进行移动

## 在这里，AdaGrad算法根据自变量在每个维度的梯度值大小来调整各个维度的学习率．
# 从而有效解决了单个学习率难以兼顾所有维度自变量梯度更新不一致的问题．
```

```python
### 具体算法
```

$$ \boldsymbol{s}_t \leftarrow \boldsymbol{s}_{t-1} + \boldsymbol{g}_t  \odot \boldsymbol{g}_t, $$

$$\boldsymbol{x}_t \leftarrow \boldsymbol{x}_{t-1} - \frac{\eta}{\sqrt{\boldsymbol{s}_t + \epsilon}} \odot \boldsymbol{g}_t,$$

这里的开方，除法和乘法运算都是按元素进行运算，因此使得目标函数自变量中每个元素都分别拥有自己的学习率．上式有如下特点：

* 由于小批量随机梯度按元素进行相乘运算，且不断累加，故$\boldsymbol{s}_t$的值在不断变大.而它正好出现在学习率的分母项，故整体学习率在不断变小/不变的．

* 如果目标函数有关自变量中某个元素的偏导数一直在较大的水平，那么该元素的学习率将下降较快，反之则较慢．

* 当学习率在早期下降得过快且尚未达到一个较佳的解时，该算法在迭代后期由于学习率较小，可能比较难以找到一个有用的解．

```python
import math
import torch
import sys
sys.path.append('../d2lzh/')
import d2lzh_pytorch as d2l

def adagrad_2d(x1, x2, s1, s2):   # 优化方法
    g1, g2, eps = 0.2 * x1, 4 * x2, 1e-6
    s1 += g1 ** 2
    s2 += g2 ** 2
    x1 -= eta / math.sqrt(s1 + eps) * g1
    x2 -= eta / math.sqrt(s2 + eps) * g2
    return x1, x2, s1, s2

def f_2d(x1, x2):
    return 0.1 * x1 ** 2 + 2 * x2 ** 2  # function

eta = 0.4   # 与上节保持一致
d2l.show_trace_2d(f_2d, d2l.train_2d(adagrad_2d))
```

```python
# 从上图可以看出什么? １)自变量移动轨迹平滑; 2)到了后期，自变量移动幅度较小．
# 原因可能是$\boldsymbol{s}_t$的不断累加使学习率不断衰减．

# 将学习率增加，同样可以看到后期自变量移动幅度较小．但前期移动幅度较大．
```

```python
eta = 3.75
d2l.show_trace_2d(f_2d, d2l.train_2d(adagrad_2d))
```

```python
### 从零开始实现
```

```python
features, labels = d2l.get_data_ch7()

def init_adagrad_states():
    s_w = torch.zeros((features.shape[1], 1), dtype=torch.float32)
    s_b = torch.zeros(1, dtype=torch.float32)
    return (s_w, s_b)

def adagrad(params, states, hyperparams):
    eps = 1e-6
    for p, s in zip(params, states):
        s.data += (p.grad.data ** 2)
        p.data -= hyperparams["lr"] * p.grad.data / torch.sqrt(s.data + eps)
```

```python
d2l.train_ch7(adagrad, init_adagrad_states(), {'lr':0.1}, features, labels)
```

```python
### 简洁实现
```

```python
d2l.train_pytorch_ch7(torch.optim.Adagrad, {'lr':0.1}, features, labels)
```

### 小结

* AdaGrad算法在迭代过程中不断调整学习率，并让目标函数自变量中每个元素都分别拥有自己的学习率。

* 使用AdaGrad算法时，自变量中每个元素的学习率在迭代过程中一直在降低（或不变）。
