```python
# RMSProp算法

# 该算法是对AdaGrad的改进．在哪方面进行了改进？
```

 原来的AdaGrad状态变量是小批量随机梯度按元素求平方的不断累积，故对应地学习率会不断降低．如果学习率已经降低到一定程度而函数距离最优解还有一段较大的距离，此时该方案将有可能失效．因此，RMSProp提出，对状态变量综合了指数加权移动平均的优点，将小批量随机梯度的先前状态纳入考虑．即

$$\boldsymbol{s}_t \leftarrow \gamma \boldsymbol{s}_{t-1} + (1 - \gamma) \boldsymbol{g}_t \odot \boldsymbol{g}_t, $$

$$\boldsymbol{x}_t \leftarrow \boldsymbol{x}_{t-1} - \frac{\eta}{\sqrt{\boldsymbol{s}_t + \epsilon}} \odot \boldsymbol{g}_t, $$

因为RMSProp算法的状态变量$\boldsymbol{s}_t$是对平方项$\boldsymbol{g}_t \odot \boldsymbol{g}_t$的指数加权移动平均，所以可以看作是最近$1/(1-\gamma)$个时间步的小批量随机梯度平方项的加权平均。如此一来，自变量每个元素的学习率在迭代过程中就不再一直降低（或不变）。

```python
### 例子
```

```python
import math
import torch
import sys
sys.path.append('../d2lzh/')
import d2lzh_pytorch as d2l

def rmsprop_2d(x1, x2, s1, s2):
    g1, g2, eps = 0.2 * x1, 4 * x2, 1e-6
    s1 = gamma * s1 + (1 - gamma) * g1 ** 2
    s2 = gamma * s2 + (1 - gamma) * g2 ** 2
    x1 -= eta / math.sqrt(s1 + eps) * g1
    x2 -= eta / math.sqrt(s2 + eps) * g2
    return x1, x2, s1, s2

def f_2d(x1, x2):
    return 0.1 * x1 ** 2 + 2 * x2 ** 2

eta, gamma = 0.2, 0.99
d2l.show_trace_2d(f_2d, d2l.train_2d(rmsprop_2d))
```

```python
### 从零开始实现
```

```python
features, labels = d2l.get_data_ch7() # 获取数据

def init_rmsprop_states():　# 初始化权重参数
    s_w = torch.zeros((features.shape[1], 1), dtype=torch.float32)
    s_b = torch.zeros(1, dtype=torch.float32)
    return (s_w, s_b)

def rmsprop(params, states, hyperparams):  # 如何更新权值
    gamma, eps = hyperparams['gamma'], 1e-6
    for p, s in zip(params, states):
        s.data = gamma * s.data + (1 - gamma) * (p.grad.data) ** 2
        p.data -= hyperparams['lr'] / torch.sqrt(s + eps) * p.grad.data
```

```python
d2l.train_ch7(rmsprop, init_rmsprop_states(), 
             {'lr':0.01, 'gamma':0.9}, features, labels)
```
